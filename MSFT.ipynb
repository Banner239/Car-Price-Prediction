Importing Libraries and Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("MSFT.csv")
print(data.head())
```

Plotting Close Prices

```python
plt.figure(figsize=(6, 4))
plt.title("Microsoft Stock Prices")
plt.xlabel("Date")
plt.ylabel("Close")
plt.plot(data["Close"])
plt.show()
```

Splitting Our Data into Training and Testing Datasets

```python
x = data[["Open", "High", "Low"]]
y = data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

Training Our Model Using Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
data = pd.DataFrame(data={"Predicted Rate": ypred})
print(data.head())
```

Plotting Our Predicted Price

```python
plt.figure(figsize=(10,6))
plt.style.use("fivethirtyeight")
plt.title("Microsoft Stock Price Prediction")
plt.xlabel("Days")
plt.ylabel("Close Price")
plt.plot(ytest, color="purple", linewidth=2, markersize=10, label="Actual Price")
plt.plot(ypred, color="green", linewidth=2, markersize=10, label="Predicted Price")
plt.legend()
plt.show()



```
